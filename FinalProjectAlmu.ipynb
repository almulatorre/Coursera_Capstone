**METHODOLOGY**  
Our analytic approach has two phases, so we can decide easily which city, and where inside the city center, has the best investment opportunity.  
In the first phase, we see the current status of both cities, related to hotels distribution in each city center. We achieve this by plotting the maps of each city, with hotels superimposed on top.  
Once we know which city has more free-of-hotel space in its central zone, in the second phase, we use the unsupervised learning technique named Clustering, so that we can choose the best zone inside city center to build a new hotel.

**Second phase**  
We use the Clustering technique named KMEANS, so that we can choose the best zone inside Madird downtown to build a new hotel.  

```python
from sklearn.cluster import KMeans
```

Let's see how many areas we have in city center.

```python
df3 = df1.groupby(['Area'])['name'].apply(lambda x: ','.join(x.astype(str))).reset_index()
df3
```

```python
#There are 9 areas, so let's take 9 clusters.
kclusters = 9
k_means = KMeans(init = "k-means++", n_clusters = kclusters, n_init = 12)
```

```python
df_cluster = df1.drop('name', 1)
k_means.fit(df_cluster)
```

```python
k_means.labels_[0:50]
```

```python
# add clustering labels
df['Cluster Labels'] = k_means.labels_
df.head()
```

We create the map

```python
import numpy as np 
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
map_cluster = folium.Map(location=[latitudeM, longitudeM], zoom_start=13)
map_cluster
```

```python
dfc = df[['Area', 'name', 'Latitude', 'Longitude', 'Cluster Labels']]
dfc.head()
```

```python
# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add hotels to the map
markers_colors = []
for lat, lon, name, cluster in zip(dfc['Latitude'], dfc['Longitude'], dfc['name'], dfc['Cluster Labels']):
    label = folium.Popup(str(name) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=10,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_cluster)
       
map_cluster
```

**RESULTS Second Phase**  
Looking at the clustering image, we can say the best zone to build a new hotel in Madrid center is the dark orange zone, that is, near "Barrio de la Latina" and "Lavapiés", as it is the area with less hotels.
